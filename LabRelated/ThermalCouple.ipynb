```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ThermalCoupleVolt =lambda temp,slope,b:temp*slope+b
```

```python
Temp=np.arange(1,300,5)
Voltage=ThermalCoupleVolt(Temp,0.2/5,1.1)  # slope 0.2mv/5 C
```

```python
plt.title("Thermal Couple Voltage to Temperature Relation")
plt.xlabel("Voltage (mV)")
plt.ylabel("Temp (C)")
plt.axvline(10.3,0,230,color="r")
plt.axhline(230,0,10.3,color="r")
plt.scatter(Voltage,Temp)
```
