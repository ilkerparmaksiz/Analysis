```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
%matplotlib inline
```

```python
#Current Avg Log.txt
def GetValues(FilePath,folders,fileName,Intensity=[],DarkCurrent={}):
    values={}
    if(len(Intensity)==0):
        for i in folders:
            File=FilePath+"/"+str(i)+"/"+fileName
            with open(File) as f:
                values[i]=abs(float(f.readline().split(" ")[0].strip()))
                f.close()
    else:
        for i in folders:
            Currents=np.array([])
            for k in Intensity:
                File=FilePath+"/"+str(i)+"/"+k+"/"+fileName
                with open(File) as f:
                    if(len(DarkCurrent)==0):
                        Currents=np.append(Currents,abs(float(f.readline().split(" ")[0].strip())))
                    else:
                        Currents=np.append(Currents,(abs(float(f.readline().split(" ")[0].strip())))-DarkCurrent[i])

                    f.close()
            values[i]=Currents

    return values

### Gain 
gainv2=lambda I,NPulse,TimeWidth,Npe,e:(I*NPulse*TimeWidth)/(Npe*e)
gain=lambda I,NPulse,TimeWidth,Npe,e:(I)/(Npe*e*NPulse)


def PlotableArray(Intensity,Voltage,d):
    tempDict={}
    for i in Voltage:
        tempDict[i]=d[i][Intensity]
    return tempDict
        
def PlotLightCurrents(values,title,DataSheet=[]):
    labels=["10Khz","100Khz","1M"]
    plt.title(title)
    plt.ylabel("Current(A)")
    plt.xlabel("Voltage(V)")
    ####
    print("Plotting Current vs Voltage....")
    for i in range(0,3):
        k=PlotableArray(i,Voltages,values) # 0 for 10k, 1 for 100k, 2 for 1M
        plt.scatter(k.keys(),k.values(),label=labels[i])
        plt.plot(k.keys(),k.values())
    plt.semilogy()
    plt.legend()
    plt.show()
    
    Intensity=[10e3,100e3,1000e3]
    gain2d=[[0],[0],[0],[0]]
    
    print("Plotting Gains...")
    count=0
    for volt in Voltages:
        gain1d=[0,0,0]
        for i in range(0,3):
            k=PlotableArray(i,Voltages,values) # 0 for 10k, 1 for 100k, 2 for 1M
            g=gain(k[volt],Intensity[i],100e-12,1,1.602e-19)
            gain1d[i]=g
            print(f"Gain is {g} at {volt} V and Intesity I= {Intensity[i]},Current={k[volt]}")
            
        gain2d[count]=gain1d 
        count=count+1


    averageGain=np.array([])
    gainError=np.array([])
    for i in gain2d:
        averageGain=np.append(averageGain,st.mean(i))
        gainError=np.append(gainError,st.stdev(i))
    print("------Average Gain------")    
    print(averageGain)
    print("------Gain Error------")    
    print(gainError)    

    plt.title(title+"_Gain")
    plt.xlabel("Voltages(V)")
    plt.ylabel("Gains")

    plt.plot(Voltages,averageGain,color='red')
    plt.scatter(Voltages,averageGain,label=title,color='red')
    plt.errorbar(Voltages, averageGain,yerr = gainError,color='red')
    if(DataSheet):
        plt.plot(Voltages,DataSheet)
        plt.scatter(Voltages,DataSheet,label="DataSheet")
        

    plt.semilogy()
    plt.legend()
    plt.show()

    
```

```python
################ Data Sheet Values ###########################
DataSheetGain=[319354.84129089303,780638.8166807669,1908212.694210148,4702888.39292543]
DataSheetDarkCurrent=[4.888652745113848e-10,1.1645150146158026e-9, 2.79677277587171e-9, 6.662129751855543e-9]

```

```python
### Dark Current Measurement 
Voltages=np.arange(900,1300,100)

DarkCurrentPMt1=GetValues("Dark_Rate/maskwith_hole_tubewithoutteflon",Voltages,"Current Avg Log.txt")
DarkCurrentPMt2=GetValues("Dark_Rate/with_mask",Voltages,"Current Avg Log.txt")
DarkCurrentPMt3=GetValues("Dark_Rate/hamamatsuPMT_R7378A_tube3",Voltages,"Current Avg Log.txt")

plt.title("Comparing Dark Currents")
plt.xlabel("Voltages (V)")
plt.ylabel("Current (A)")

plt.plot(DarkCurrentPMt1.keys(),DarkCurrentPMt1.values(),color="red")
plt.scatter(DarkCurrentPMt1.keys(),DarkCurrentPMt1.values(),label="PMT1",color="red")
plt.plot(DarkCurrentPMt2.keys(),DarkCurrentPMt2.values(),color="blue")
plt.scatter(DarkCurrentPMt2.keys(),DarkCurrentPMt2.values(),label="PMT2",color="blue")
plt.plot(DarkCurrentPMt3.keys(),DarkCurrentPMt3.values(),color="green")
plt.scatter(DarkCurrentPMt3.keys(),DarkCurrentPMt3.values(),label="PMT3",color="green")
plt.plot(Voltages,DataSheetDarkCurrent,color="purple")
plt.scatter(Voltages,DataSheetDarkCurrent,label="DataSheet",color="purple")
plt.semilogy()
plt.legend()

print("PMT_1")
print(DarkCurrentPMt1)

print("PMT_2")
print(DarkCurrentPMt2)

print("PMT_3")
print(DarkCurrentPMt3)

```

```python
### Light Measurement
Intensities=["10k","100k","1M"]
Light_PMT1_values=GetValues("with light/tube_withoutteflon",Voltages,"Current Avg Log.txt",Intensities,DarkCurrentPMt1)
Light_PMT2_values=GetValues("with light/tube_withteflon",Voltages,"Current Avg Log.txt",Intensities,DarkCurrentPMt2)
Light_PMT3_values=GetValues("Dark_Rate/hamamatsuPMT_R7378A_tube3",Voltages,"Current Avg Log.txt",Intensities,DarkCurrentPMt3)
```
