```python
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
%matplotlib inline
```

```python
#Current Avg Log.txt
## It grabs the values for a given file and places it into a dictionary
def GetValues(FilePath,folders,fileName,SubFolders=[]):
    values={}
    Files=np.array([])   
    for i in folders:
        if(len(SubFolders)==0):
            File=FilePath+"/"+str(i)+"/"+fileName
            Files=np.append(Files,File)
        else:
            for Sub in SubFolders:
                if (type(Sub) is not str):
                    Sub=str(Sub)
                File=FilePath+"/"+str(i)+"/"+Sub+"/"+fileName
                Files=np.append(Files,File)
        print(f"Opening {len(Files)} many Files to analyze")
        print(Files)                        

        for F in Files:
            Rawdata=np.array([])                    
            with open(F) as f:
                Lines=f.readlines()
                TotalLines=len(Lines)
                if(TotalLines==0):
                    print("File is Empty!")
                elif(TotalLines==1):
                    values[i]=abs(float(Lines[0].split(" ")[0].strip()))
                elif(TotalLines==2):
                    values[i]=abs(float(Lines[1].split(" ")[0].strip()))
                else:
                    for line in Lines:
                        Rawdata=np.append(Rawdata,abs(float(line.split(" ")[0].strip())))
            f.close()
            
            AveragedData=AverageIt(Rawdata)
            values[i]=AveragedData
    return values

### Gain Calculations
gainv2=lambda I,NPulse,TimeWidth,Npe,e:(I*NPulse*TimeWidth)/(Npe*e)
gain=lambda I,NPulse,TimeWidth,Npe,e:(I)/(Npe*e*NPulse)



### This needs better explanation
#### this creates a dictionary from a given Intesity 
def CreateADict(Intensity,Voltage,data):
    tempDict={}
    for i in Voltage:
        tempDict[i]=data[i][Intensity]
    return tempDict


### This Function taking care of the plotting of the data obtained from the file
def PlotLightCurrents(values,title,DataSheet=[]):
    labels=["10Khz","100Khz","1Mhz"]
    plt.title(title)
    plt.ylabel("Current(A)")
    plt.xlabel("Voltage(V)")
    ####
    print("Plotting Current vs Voltage....")
    for i in range(0,3):
        k=CreateADict(i,Voltages,values) # 0 for 10k, 1 for 100k, 2 for 1M
        plt.scatter(k.keys(),k.values(),label=labels[i])
        plt.plot(k.keys(),k.values())
    plt.semilogy()
    plt.legend()
    plt.show()
    
    Intensity=[10e3,100e3,1000e3]
    gain2d=[[0],[0],[0],[0]]
    
    print("Plotting Gains...")
    count=0
    for volt in Voltages:
        gain1d=[0,0,0]
        for i in range(0,3):
            k=CreateADict(i,Voltages,values) # 0 for 10k, 1 for 100k, 2 for 1M
            g=gain(k[volt],Intensity[i],100e-12,1,1.602e-19)
            gain1d[i]=g
            print(f"Gain is {g} at {volt} V and Intesity I= {Intensity[i]},Current={k[volt]}")
            
        gain2d[count]=gain1d 
        count=count+1


    averageGain=np.array([])
    gainError=np.array([])
    for i in gain2d:
        averageGain=np.append(averageGain,st.mean(i))
        gainError=np.append(gainError,st.stdev(i))
    print("------Average Gain------")    
    print(averageGain)
    print("------Gain Error------")    
    print(gainError)    

    plt.title(title+"_Gain")
    plt.xlabel("Voltages(V)")
    plt.ylabel("Gains")

    plt.plot(Voltages,averageGain,color='red')
    plt.scatter(Voltages,averageGain,label=title,color='red')
    plt.errorbar(Voltages, averageGain,yerr = gainError,color='red')
    if(DataSheet):
        plt.plot(Voltages,DataSheet)
        plt.scatter(Voltages,DataSheet,label="DataSheet")
        

    plt.semilogy()
    plt.legend()
    plt.show()

### Helps to Plot a dictinary and a list with a linear fit    
def PlotWithTrendLines(data,lbl,clr,fit=True):
    #Calculating the trendline
    
    if(type(data) is dict):
        x=np.array(list(data.keys()))
        y=list(data.values())
    else:
        x=data[0]
        y=data[1]
        
        
    # the line equation:
    print(lbl)    
    if(fit):  
        z = np.polyfit(x,y, 1)
        p = np.poly1d(z)
        plt.plot(x,p(x),"--",color=clr)
        print(f"y= {z[0]}x + {z[1]}")
    else:
        plt.plot(x,y,label=lbl,color=clr)

    plt.scatter(x,y,label=lbl,color=clr)
   
  
    print (f" X values --> /n {x}")
    print (f" Y values --> /n {y}")
    
def AverageIt(data):
    if (len(data)==0):
        print("Given Data is empty")
    else:
        if(type(data) is list):
            data=np.array(data)
            
        mean=data.sum()/len(data)
        x=(data-mean)**2
        stdev=np.sqrt(x.sum()/(len(x)-1))
        
    return mean,stdev

def SubstractNoise(data,noise):
    newDict={}
    value=[0,0]
    for key in data:
        value[0]=data[key][0]-noise[key][0]  # Substracting the Means
        value[1]=np.sqrt(data[key][1]**2+noise[key][1]**2) ## Error Propagating to get the new Standart Deviation
        newDict[key]=value
        
    return value
        
    
```
